파일 저장 불러오기

```python
import numpy as np
# delimiter 구분자 (공백, csv파일은 ',')
# 저장

score = [[73., 80., 75., 152.],
        [93., 88., 93., 185.],
        [89., 91., 90., 180.],
        [96., 98., 100., 196.],
        [73., 66., 70., 142.],
        [53., 46., 55., 101.]]


np.savetxt('savetxt.csv', score,
           header='EXAM1, EXAM2, EXAM3, FINAL',
           delimiter=', ')
```

```python
# 불러오기

xy = np.loadtxt('savetxt.csv', delimiter=',', dtype=np.float32)
x_data = xy[:, 0:-1]
y_data = xy[:, [-1]]

print(x_data.shape, '\n', x_data, '\n')
print(y_data.shape, '\n', y_data, '\n')
```

```python
import numpy as np
import tensorflow as tf

tf.set_random_seed(777)

xy = np.loadtxt('savetxt.csv', delimiter=',', dtype=np.float32)
x_data = xy[:, 0:-1]
y_data = xy[:, [-1]]

# print(x_data.shape, '\n', x_data, '\n')
# print(y_data.shape, '\n', y_data, '\n')

x = tf.placeholder(tf.float32, shape=[None, 3])
y = tf.placeholder(tf.float32, shape=[None, 1])

w = tf.Variable(tf.random_normal([3, 1]), name = "weight")
b = tf.Variable(tf.random_normal([1]), name = "bias")

hypothesis = tf.matmul(x, w) + b

cost = tf.reduce_mean(tf.square(hypothesis - y))


optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)

sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(2001):
    cost_val, hy_val, _ = sess.run([cost, hypothesis, train],
                                  feed_dict = {x: x_data, y: y_data})
    if step % 100 == 0:
        print(step, "Cost : " , cost_val, "\n Prediction: \n", hy_val, '\n')
      
    
# 2차원이니까 2차원배열로
print("Your score will be", sess.run(hypothesis, feed_dict={x: [[100, 70, 101]]}))
print("Your score will be", sess.run(hypothesis, feed_dict={x: [[60, 70, 110], [90, 100, 80]]}))
```

# 여러 파일 불러오기
1. 파일네임 리스트화
2. 파일 읽어올 reader 정의
3. 불러온 값에 어떻게 파싱할지 (record_defaults=데이터타입 정해줌 +
                              decode_csv)
4. 배치 (batch_size=몇개 가져올지)
5. 큐에넣기(일반적으로 선언)
6. 호출

-셔플배치는 참고


```python

tf.reset_default_graph()

score = [[73., 80., 75., 152.],
        [93., 88., 93., 185.],
        [89., 91., 90., 180.],
        [96., 98., 100., 196.],
        [73., 66., 70., 142.],
        [53., 46., 55., 101.]]


np.savetxt('savetxt.csv', score,
           delimiter=', ')


# 파일 네임 리스트화
filename_queue = tf.train.string_input_producer(
    ['savetxt.csv'], shuffle=False, name='filename_queue')


# 파일 읽어올 reader정의
reader = tf.TextLineReader()
key, value = reader.read(filename_queue)

# 불러온 값에 어떻게 파싱할지
# record_defaults = 데이터 타입 정해줌 + decode_csv
record_defaults = [[0.], [0.], [0.], [0.]]
xy = tf.decode_csv(value, record_defaults=record_defaults)

# 받아옴 batch_size=한번에 몇개가져올지
train_x_batch, train_y_batch = \
tf.train.batch([xy[0:-1], xy[-1:]], batch_size=10)

x = tf.placeholder(tf.float32, shape=[None, 3])
y = tf.placeholder(tf.float32, shape=[None, 1])



w = tf.Variable(tf.random_normal([3, 1]), name = "weight")
b = tf.Variable(tf.random_normal([1]), name = "bias")

hypothesis = tf.matmul(x, w) + b

cost = tf.reduce_mean(tf.square(hypothesis - y))


optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)
   
sess = tf.Session()
sess.run(tf.global_variables_initializer())


######
coord = tf.train.Coordinator()
threads = tf.train.start_queue_runners(sess=sess, coord=coord)


for step in range(2001):
    x_batch, y_batch = sess.run([train_x_batch, train_y_batch])
    cost_val, hy_val, _ = sess.run([cost, hypothesis, train],
                                   feed_dict = {x: x_batch, y: y_batch})
    if step % 100 == 0:
        print(step, "Cost : " , cost_val, "\n Prediction: \n", hy_val)
 
coord.request_stop()
coord.join(threads)
```
